```python
#importing required Libraries
import pandas as pd   
import requests       
import re
from bs4 import BeautifulSoup 
import numpy as np  
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns


# to bypass warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#assigning the URL with variable name url
url = 'https://www.imdb.com/search/title/?count=250&groups=top_1000&sort=user_rating'

#request allow one to send HTTP request
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup)
```

```python
#creating an empty list, so that i can append the values
movie_title = []  
year_of_release = []  
ratings = []  
duration = [] 
genre = []  
metascore = []  
votes = []  
movie_description = [] 
director = []  
stars = []  
gross = []
```

```python
#storing the required data in the variable
movie_data = soup.findAll('div', attrs= {'class': 'lister-item mode-advanced'})
```

```python
movie_data
```

```python
print(len(movie_data))
```

```python
#Extracting the dataset using for loop
for x in movie_data:
    name = x.h3.a.text
    movie_title.append(name)
    
    year= x.h3.find('span',class_='lister-item-year text-muted unbold').text.replace('(', '').replace(')','')
    year_of_release.append(year)
    
    rating = x.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n', '')
    ratings.append(rating)
    
    movie_time= x.p.find('span', class_ = 'runtime').text
    duration.append(movie_time)
    
    movie_genre=x.p.find('span', class_ = 'genre').text.replace('\n', '')
    genre.append(movie_genre)
    
    #Some of the metascore might not be missing so use the if statement to return those missing
    meta = x.find('span', class_ = "metascore").text if x.find('span', class_ = "metascore") else "0"
    metascore.append(meta)
    
    description = x.find_all('p', class_ = 'text-muted')
    description_= description[1].text.replace('\n', '') if len(description) >1 else 'not available'
    movie_description.append(description_)
    
   #Votes and gross have same attributes, so i created a common variable and then used indexing
    value = x.find_all('span', attrs = {'name': 'nv'})
    
    vote = value[0].text
    votes.append(vote)

    movie_gross = value[1].text if len(value) >1 else '0'
    gross.append(movie_gross)
    
    #Details of casts -- Scraping Director name and Stars
    cast = x.find("p", class_ = '').text.replace('\n', '').split('|')
    cast = [x.strip() for x in cast]
    cast = [cast[i].replace(j, "") for i,j in enumerate(["Director:", "Stars:"])]
    director.append(cast[0])
    stars.append([x.strip() for x in cast[1].split(",")])
```

```python
#creating a dataframe using pandas library
movies= pd.DataFrame({'Movie_Title': movie_title, 'Year_of_Release': year_of_release,'Movie_Rating': ratings, 'Movie_Duration':duration,'Movie_Genre':genre,'Metascore':metascore,'Movie_Votes':votes,
                         'Movie_Description':movie_description,'Movie_Director':director,'Movie_Stars':stars,'Gross':gross})   
```

### DATA CLEANING 

```python
##To see the first genre for the top 250movies
movies['Movie_Genre'].str.split(',',expand=True)
```

```python
#To reassign the first genre of the movie to movie Genre column
movies['Movie_Genre']=movies['Movie_Genre'].str.split(',',expand=True)[0]
```

```python
movies.Movie_Genre.value_counts()
```

```python
#To remove trailing spaces from each column and convert the datatypes
movies['Movie_Genre'] = movies['Movie_Genre'].str.replace(' ', '')
movies['Metascore'] = movies['Metascore'].str.replace(' ', '').astype(int)
movies['Gross'] = movies['Gross'].str.replace(' ','').str.replace('$', '').str.replace('M', '').str.replace('#','').astype(float)
movies['Movie_Votes'] = movies['Movie_Votes'].str.replace(' ','').str.replace(',', '').astype(int)
movies['Movie_Rating'] = movies['Movie_Rating'].str.replace(' ', '')
movies['Year_of_Release'] = movies['Year_of_Release'].str.replace(' ', '')
movies['Movie_Duration'] = movies['Movie_Duration'].str.replace(' min', '').astype(float)
```
